```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np   
from scipy.integrate import odeint

R=8.3145        ##Joules/mole
G=6.674e-11     ## m^3/(kg*s^2)
X=0.64
kappa=.1
epsi0=1.08e-12  ## W*m^3/kg^2
c=3e8           ##speed of light m/s
M_sun=2.0e30  
L_sun=3.8e26
R_sun=7.0e8
a=1

params=R,G,X,kappa,epsi0,c,M_sun,L_sun,R_sun

def rhofunc(P,T,R):
    return P/(R*T)

def epsi(rho,T,epsi0):
    return epsi0*rho*np.power(X,2)*np.power((T/10**6),4)

def dM_dr(r,rho):
    dMdr=4.0*np.pi*np.power(r,2)*rho
    return dMdr

def dP_dr(r,Mr,rho,G):
    dPdr=-G*Mr*rho/np.power(r,2)
    return dPdr

def dL_dr(r,rho,T,epsi0):
    epsi2=epsi(rho,T,epsi0)
    dLdr=4.0*np.pi*np.power(r,2)*rho*epsi2
    return dLdr

def dT_dr(T,r,rho,Lr,kappa,a,c):
    dTdr=-3.0/(4.0*a*c)*kappa*rho/np.power(T,3)*Lr/(4.0*np.pi*np.power(r,2))
    return dTdr

def butts(x,y):
    return(x-y)




def Mother_function(y,r,params):
    R,G,X,kappa,epsi0,c,M_sun,L_sun,R_sun,a,c=params
    Mr,P,Lr,T=y
    rho=rhofunc(P,T,R)
    derivs = [dM_dr(r,rho), dP_dr(r,Mr,rho,G) , dL_dr(r,rho,T,epsi0) , dT_dr(T,r,rho,Lr,kappa,a,c)]
    return derivs


##Parameters
params=R,G,X,kappa,epsi0,c,M_sun,L_sun,R_sun,a,c

##  initial conditions at r=R_sun
M0=0
L0=0
T0=1.57e6   # Kelvin
P0=2.3e16  # kg/(m*s^2)
r0=0.0001
rho=rhofunc(P0,T0,R)


y0=[M0, P0, L0, T0]


r=np.linspace(0.0001,7e8,100)

Mother_function(y0,r,params)

solution=odeint(Mother_function,y0,r,args=(params,))
```
